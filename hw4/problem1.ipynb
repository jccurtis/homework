# Begin Homework

```python
from os import listdir
from multiprocessing import Pool, cpu_count
from pylab import imread
from time import time
import skimage
import numpy as np

MYDIRECTORY = "/Users/josephcurtis/python/ay250/homework/hw4/50_categories"

# Quick function to divide up a large list into multiple small lists, 
# attempting to keep them all the same size. Author: Christopher Klein, Joshua Bloom
def split_seq(seq, size):
        newseq = []
        splitsize = 1.0/size*len(seq)
        for i in range(size):
            newseq.append(seq[int(round(i*splitsize)):
                int(round((i+1)*splitsize))])
        return newseq

# FEATURE EXTRACTION
def extract_features(im_cat_path_list):
    feature_list = []
    for im_cat, im_path in im_cat_path_list:
        im_array = imread(im_path) #image matrix
        feature1 = im_array.size #image size
        #print shape(im_array)
        #feature2 = im_array.
        feature_list.append([im_cat, im_path, feature1])
    return feature_list

# LIST OF LISTS STORING CATAGORY AND IMAGE PATH
im_cat_paths = []
categories = listdir(MYDIRECTORY)
for category in categories:
    im_names = listdir(MYDIRECTORY  + "/" + category)
    for name in im_names:
        im_cat_paths.append([category, MYDIRECTORY + "/" + category + "/" + name])
        
# CHECK IMAGES
print ("There should be 4244 images, actual number is " + str(len(im_cat_paths)) + ".")

# RUN ON ALL CORES
numprocessors = cpu_count()

# DEFINE LIST OF SMALLER LISTS
split_im_cat_paths = split_seq(im_cat_paths, numprocessors)

# RUN FEATURE EXTRACTION IN PARALLEL
start_time = time()
p = Pool(numprocessors)
result = p.map_async(extract_features, split_im_cat_paths)
poolresult = result.get()
end_time = time()

# PRINT RESULTS
print ("Finished extracting features. Total time: " + str(round(end_time-start_time, 3)) + " s, or " + str(round((end_time-start_time)/len(im_cat_paths), 5)) + " s/image.")

# PUT DIFFERENT CORE RESULTS INTO ONE LIST
combined_result = []
for single_proc_result in poolresult:
    for single_im_result in single_proc_result:
        combined_result.append(single_im_result)

# RANDOMIZE LIST
np.random.shuffle(combined_result)

# FEATURES and CATAGORIES 
rf_features = [];
for result in combined_result: rf_features.append(result[2:])
rf_categories = [];
for result in combined_result: rf_categories.append(result[0])

# TEST AND TRAINING SET (HALF)
rf_features_train = rf_features[:4244/2]
rf_features_test = rf_features[4244/2:]
rf_categories_train = rf_categories[:4244/2]
rf_categories_test = rf_categories[4244/2:]

# NUMPY ARRAYS
np.array(rf_features,dtype=float)
np.array(rf_features_train,dtype=float)
np.array(rf_features_test,dtype=float)
np.array(rf_categories, dtype=str)
np.array(rf_categories_train, dtype=str)
np.array(rf_categories_test, dtype=str)
```

#Make Classifier

```python
from sklearn.ensemble import RandomForestClassifier

# random forest classifier with 1000 trees and the default split value = sqrt(15)
rfc = RandomForestClassifier(n_estimators=1000, n_jobs=-1)
# random forest classifier fitting to classifiers and targets
rfc.fit(rf_features_train,rf_categories_train)
```

#Test Classifier

```python
from sklearn import metrics

predicted_catagories = rfc.predict(rf_features_test) #prediction
rfc_confmat = metrics.confusion_matrix(rf_categories_test, predicted_catagories) #confusion matrix
rfc_accuracy = metrics.accuracy_score(rf_categories_test, predicted_catagories) # zero-one score
print("Zero-One (Accuracy) Score: {0:0.2f}%".format(rfc_accuracy*100))
```

#Sample Corner Detection

```python
%matplotlib inline
from matplotlib import pyplot as plt

from skimage import data
from skimage.feature import corner_harris, corner_subpix, corner_peaks
from skimage.transform import warp, AffineTransform
from skimage.draw import ellipse

tform = AffineTransform(scale=(1.3, 1.1), rotation=1, shear=0.7,
                        translation=(210, 50))
image = warp(data.checkerboard(), tform.inverse, output_shape=(350, 350))
rr, cc = ellipse(310, 175, 10, 100)
image[rr, cc] = 1
image[180:230, 10:60] = 1
image[230:280, 60:110] = 1

coords = corner_peaks(corner_harris(image), min_distance=5)
coords_subpix = corner_subpix(image, coords, window_size=13)

plt.gray()
plt.imshow(image, interpolation='nearest')
plt.plot(coords[:, 1], coords[:, 0], '.b', markersize=3)
plt.plot(coords_subpix[:, 1], coords_subpix[:, 0], '+r', markersize=15)
plt.axis((0, 350, 350, 0))
plt.show()
```
