# Pandas dataframe import

The given CSV files were imported with Pandas dataframes. They are joined to form the airports dataframe which is a left join on topAirports with ICAOAirports.

````python
topAirports = pd.io.parsers.read_csv('hw_6_data/top_airports.csv')
ICAOAirports = pd.io.parsers.read_csv('hw_6_data/ICAO_airports.csv')
airports = topAirports.join(ICAOAirports)
````

# Weather Underground Scraping Code

This code was run to build the 50 csv files (one for each top 50 airport) in the `weather_data` directory. Each csv file is a combined list of all the weather data from 1/1/2008 to 10/15/2013 (when the code was ran).

```python
startDate = dt.datetime(2008,1,1)
endDate = dt.datetime.now()

#columns for dataframes
colList = ['Date', 
            'Max TemperatureF', 
            'Mean TemperatureF',

for ICAO in airports.ICAO:
    for year in range(startDate.year,endDate.year+1):
        for month in range(startDate.month,13):
            if year < endDate.year: #so the remaining (future) months in the current year aren't counted
                url = 'http://www.wunderground.com/history/airport/{:s}/{:d}/{:d}/1/MonthlyHistory.html?format=1'.format(ICAO,year,month)
            elif month <= endDate.month: #so the remaining (future) months in the current year aren't counted
                url = 'http://www.wunderground.com/history/airport/{:s}/{:d}/{:d}/1/MonthlyHistory.html?format=1'.format(ICAO,year,month)
            else: #if current year and past current month do nothing
                break
            if year == startDate.year and month == startDate.month: #if first date for ICAO airport create dataframe
                airportDataFrame = pd.read_csv(url,comment='<',skiprows=1)
                airportDataFrame.columns = colList
            else: #if NOT first date for ICAO airport append to dataframe to make long list of all data organized by date
                airportDataFrameTEMP = pd.read_csv(url,comment='<',skiprows=1)
                airportDataFrameTEMP.columns = colList
                airportDataFrame = airportDataFrame.append(airportDataFrameTEMP,ignore_index=True)
    airportDataFrame.to_csv('weather_data/{:s}.csv'.format(ICAO))
    print('Wrote weather_data/{:s}.csv to disk'.format(ICAO))
```

#List of all dates

- I did some reformatting in the date string because the date string from Weather Underground has no leading zeros on the month and day.
- The list has 2115 entries which I use in the following step to make sure the scraped data isn't missing any days.

````python
startDate = dt.datetime(2008,1,1)
endDate = dt.datetime(2013,10,15)
delta = endDate - startDate
dateList = []
for i in range(delta.days + 1):
    dateTemp = startDate + dt.timedelta(days=i)
    dateList.append(dateTemp.strftime('%Y-%m-%d'))
dateListReformat = []
for date in dateList:
     dateListReformat.append(date.replace('-0','-'))

print('The required number of days is {:d}.'.format(len(dateList)))
````

#Load dictionary of data frames containing scraped data CSV files

````python
weatherDict = {}
for airport in airports.ICAO:
    weatherDict[airport] = pd.read_csv('weather_data/{:s}.csv'.format(airport))
    print('Loaded weather_data/{:s}.csv'.format(airport))
````

#Check Scraped Data for Completeness

Do all data sets have the required number of days (**2115**) til October 15th, 2013? - **YES**

````python
for airport in weatherDict:
    if len(weatherDict[airport].index) != len(dateList):
        print("Airport {:s} has {:d} dates instead of {:d} dates.".format(airport,len(weatherDict[airport].index),len(dateList)))
````

All dates present. Is there **max temperature** or **cloud cover** data missing? If so: interpolate (linear with each index being equal spacing, 1 day).

````python
for airport in weatherDict:
    numNullT = sum(weatherDict[airport]['Max TemperatureF'].isnull())
    numNullC = sum(weatherDict[airport]['CloudCover'].isnull())
    if numNullT > 0:
        print('{:s} was missing {:d} max temp points'.format(airport,numNullT))
        weatherDict[airport]['Max TemperatureF'] = weatherDict[airport]['Max TemperatureF'].interpolate()
    if numNullC > 0:
        print('{:s} was missing {:d} cloud cover points'.format(airport,numNullC))
        weatherDict[airport]['CloudCover'] = weatherDict[airport]['CloudCover'].interpolate()
````
