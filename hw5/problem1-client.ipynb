```python
from __future__ import print_function
import xmlrpclib, sys
%pylab inline

class image_client(object):
    
    def __init__(self, host, port):
        self.server = xmlrpclib.ServerProxy("http://{:s}:{:d}".format(host, port))
        self.available_methods = self.server.system.listMethods()
        self.im_list = []
        self.im_array = array([])
        
    def get_server_methods(self):
        print("Available methods from server:")
        for method in self.available_methods:
            print("\t",method)

    def read_image(self,image_path):
        '''
        Takes an image path as input and 
        '''
        try:
            self.im_array = imread(image_path)
        except:
            raise ValueError('The file cannot be read as an image.')
    
    def pack_image(self):
        '''
        Takes the image im_array attribute in numpy array form and 
        sets the im_list attribute as a list for xmlrpc transfer.
        '''
        self.im_list = self.im_array.tolist()
    
    def unpack_image(self):
        '''
        Take the image im_list attribute in list form from xmlrpc transfer and 
        sets the im_array attribute as a numpy array.
        '''
        self.im_array = array(self.im_list)
    
    def im_show(self):
        '''
        Input image in numpy array form and displays the image based on its color channels.
        If 1 color channel the image is displayed as greyscale. If multiple color channels
        it is displayed normally. If the array is not 2D or 3D an error is raised.
        '''
        if len(self.im_array.shape) == 3:
            imshow(self.im_array)
        elif len(self.im_array.shape) == 2:
            imshow(self.im_array,cmap=cm.Greys)
        else:
            raise ValueError('The image array to display is not of normal image dimensions.')
    
    def im_save(self, im_array):
        return None
    
    def im_blur(self, im_path):
        '''
        Input path to image. The image is imported as a numpy array, converted to a list 
        and sent to the server for image processing and sent back.
        '''
        self.read_image(im_path)
        self.pack_image()
        print('Sending image to server for processing.')
        self.im_list = self.server.im_blur(self.im_list)
        print('Processed image recieved from server.')
        self.unpack_image()
        self.im_show()

host, port = "", 5020
test_client = image_client(host,port)
```

```python
im_temp = im[1]
figure()
imshow(im_temp)
px_height = im_temp.shape[0]
px_width  = im_temp.shape[1]

px_height_half = floor(px_height/2)
px_width_half = floor(px_width/2)

im_mask = zeros_like(im_temp, dtype=float)
for y in range(px_height):
    for x in range(px_width):
        px = sqrt((px_height_half-y)**2+(px_width_half-x)**2)
        im_mask[y,x,0] = px
        im_mask[y,x,1] = px
        im_mask[y,x,2] = px

figure()
imshow(im_mask)
```
