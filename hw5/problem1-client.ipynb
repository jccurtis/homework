###Notes:

- I made a client class which communicates with the server for the user. The user interacts by initializing a image_client object and using the  three methods designed to work with the server. These methods allow the user to only input the image name (in the same directory) and the class does the rest (transforming, saving, plotting).
- I used ndimage for the blurring.
- I calculate the vignette mask using the normalized Euclidean distance magntude from the image center.
- I use many hidden methods and attributes to not clutter the users space.

```python
from __future__ import print_function
import xmlrpclib, sys
%pylab inline

class image_client(object):
    '''
    Class made to interact with its pair server class (in the neighboring notebook.)
    This class allows the user to interface with the server for image manipulation 
    easily. The user needs to initialize the server with a host and port matching 
    that of the paired server. The user has four methods:
    
        getServerMethods       - see what the server offers.
        blur                   - guassian blur with kernel 0.5% of max pixel dim. Uses ndimage
        swapColorAndDirections - moves each color channel to the next and flip left/right 
                                 orientation. If greyscale, the image is only flipped 
                                 left/right.
        vignette               - applies a color dodge vinette around the border of the 
                                 image. Uses a mask with the magnitude of the Euclidean
                                 distance weighting each original color channel (only one if 
                                 greyscale).
        
    For the imaging manipulation methods we have the following criteria:
    
        - input an image name (same directory) and the image is transmitted to the 
          server for processing.
        - modified and original images are saved on the server and client side 
          with tags appended to the original file name.
        - each method returns the resultant image numpy array and plots the array in the
          current workspace.
    
    '''
    def __init__(self, host, port):
        self.server = xmlrpclib.ServerProxy("http://{:s}:{:d}".format(host, port))
        self.availableMethods = self.server.system.listMethods()
        self._imList_ = []
        self._imArray_ = array([])
        
    def getServerMethods(self):
        print("Available methods from server:")
        for method in self.availableMethods:
            print("\t",method)

    def _imRead_(self,imName):
        '''
        Takes an image name as input, makes sure it can be read and 
        sets the imArray and imName attributes correctly.
        '''
        try:
            self._imArray_ = imread(imName)
            self._imName_ = imName
        except:
            raise ValueError('The file cannot be read as an image.')
    
    def _imPack_(self):
        '''
        Takes the image imArray attribute in numpy array form and 
        sets the imList attribute as a list for xmlrpc transfer.
        '''
        self._imList_ = self._imArray_.tolist()
        print('Saving original image as {:s}'.format('client_orig_'+self._imName_))
        imsave('client_orig_'+self._imName_,self._imArray_,format='png')
    
    def _imUnpack_(self):
        '''
        Take the image imList in list form from xmlrpc transfer and 
        sets the imArray attribute as a numpy array. The shape of the
        numpy array is stored as an attribute. imName is the image name
        from the client which is stored as an attribute and used to save
        the file on the server.
        '''
        self._imArray_ = array(self._imList_)
        self._imShape_ = self._imArray_.shape
        print('Saving modified image as {:s}'.format('client_mod_'+self._imName_))
        imsave('client_mod_'+self._imName_,self._imArray_,format='png')
    
    def _imShow_(self):
        '''
        Input image in numpy array form and displays the image based on its color channels.
        If 1 color channel the image is displayed as greyscale. If multiple color channels
        it is displayed normally. If the array is not 2D or 3D an error is raised.
        '''
        if len(self._imArray_.shape) == 3:
            imshow(self._imArray_)
        elif len(self._imArray_.shape) == 2:
            imshow(self._imArray_,cmap=cm.Greys_r)
        else:
            raise ValueError('The image array to display is not of normal image dimensions.')
    
    def blur(self, imName):
        '''
        Input path to image. The image is imported as a numpy array, converted to a list 
        and sent to the server for image processing and sent back.
        
        Performs Guassian blurring on the image with a guassian kernel of size
        0.5% of the maximum pixel dimension using the ndimage module.
        '''
        self._imRead_(imName)
        self._imPack_()
        print('Sending image to server for processing.')
        self._imList_ = self.server.blur(self._imList_,imName)
        print('Processed image recieved from server.')
        self._imUnpack_()
        self._imShow_()
        return self._imArray_
    
    def swapColorAndDirections(self, imName):
        '''
        Input path to image. The image is imported as a numpy array, converted to a list 
        and sent to the server for image processing and sent back.
        
        Shifts each color channel to the next one:
            Red   -> Green
            Green -> Blue
            Blue  -> Red
        Flips the image on the x axis (left/right).
        
        NOTE: If the image is greyscale, only the x axis is flipped 
        since there are no color channels to swap.
        '''
        self._imRead_(imName)
        self._imPack_()
        print('Sending image to server for processing.')
        self._imList_ = self.server.swapColorAndDirections(self._imList_,imName)
        print('Processed image recieved from server.')
        self._imUnpack_()
        self._imShow_()
        return self._imArray_
    
    def vignette(self, imName):
        '''
        Input path to image. The image is imported as a numpy array, converted to a list 
        and sent to the server for image processing and sent back.
        
        Performs vignetting to the image using a mask (with the center bright) 
        for pixel burning using the magnitude of the Euclidian distance from 
        the center of the image as the mask value. The mask is then normalized
        and multiplied by each color channel (only one if greyscale).
        '''
        self._imRead_(imName)
        self._imPack_()
        print('Sending image to server for processing.')
        self._imList_ = self.server.vignette(self._imList_,imName)
        print('Processed image recieved from server.')
        self._imUnpack_()
        self._imShow_()
        return self._imArray_
        
host, port = "", 5020
test_client = image_client(host,port)
#print('Connected to server at http://{:s}:{:d}'.format(host, port))
```
