```python
from __future__ import print_function
import aifc
import struct
%pylab inline

class audio_analysis(object):
    
    def __init__(self,sound_path):
        '''
        The audio_analysis class construction. The class requires a path to an uncompressed *.aiff file.
        The object is built with  a variety of attributes of this imported sound file. The number of 
        audio channels is checked and accounted for as well as the sample byte size. This assumes the 
        samples are either 2 bytes or 4 bytes. For all the audio given the samples are 2 bytes. (2 
        samples per 32 bit word.
        '''
        self.sound_path = sound_path
        self.sound_file = aifc.open(sound_path)
        self.frame_rate = self.sound_file.getframerate()
        self.num_frames = self.sound_file.getnframes() #
        self.num_channels = self.sound_file.getnchannels()
        self.frame_width = self.sound_file.getsampwidth() #number of frames per 32bit word
        self.raw_frames = self.sound_file.readframes(num_frames) #string of hex data
        self.frame_dtype = {2:'uint16',4:'uint32'} #dictionary to map number of bytes per frames per 32 words to integer data type
        self.frames_int = fromstring(raw_frames,dtype=self.frame_dtype[self.frame_width]) #numpy array of integer value of each frame
        if num_channels == 2:
            frames_int_left = self.frames_int[::2] #left channel
            frames_int_right = self.frames_int[1::2] #right channel
            self.frames_int = frames_int_left+frames_int_right #sum of left and right channel for fft analysis
        self.time = arange(size(self.frames_int)) / float(self.frame_rate) #array of timing for every time sample
        print('audio file:       {:s}'.format(self.sound_path))
        print('num_frames:      ',num_frames,'in the file')
        print('frame_rate:      ',frame_rate,'Hz')
        print('num_channels:    ',num_channels,'(1 for mono, 2 for stereo)')
        print('frame_width:     ',frame_width,'bytes')
        print('length of sound: ',time[-1],'seconds')
    
    def plot_sound(self):
        '''
        Plots the raw sound file data. For these contant tones, the pulses are high and
        steady before zeroing at the end.
        '''
        figure()
        plot(self.time,self.frames_int)
        xlabel("Time [s]")
        ylabel("Amplitude")
        title("Sound File")
    
    def fft(self):
        '''
        Performs a 1D discrete fast fourier transform on the integer audio data 
        attribute. Sets the following attributes:
            frames_fft_magn    = polar magnitude of the fft
            frames_fft_magn_dB = polar magnitude of the fft scaled in dB relative to the min
            delta_hertz        = the frequency sample spacing
            freq_hertz         = of len(frames_fft_magn) to map sample number to frequency
        '''
        #Discrete Fourier Transform of Sound
        frames_fft = fftpack.fft(self.frames_int)
        #Polar magnitude of fft complex result
        self.frames_fft_magn = sqrt(frames_fft.real**2+frames_fft.imag**2)
        #Scaling the fft magnitude to dB relative to the minimum amplitude
        self.frames_fft_magn_dB = 10*log10(frames_fft_magn/frames_fft_magn.min())
        #Mapping frequency to samples:
        #The spacing of the fft samples is equal to the inverse of the duration of the original discrete pulse. 
        #With this freq spacing we create an array with the same length as the num of samples and proper 
        #frequency spacing. This correlates the power of the fft to a frequency.
        self.delta_hertz = 1/self.time[-1] #frequency spacing
        self.freq_hertz = arange(0,len(frames_fft)*self.delta_hertz,self.delta_hertz) #frequency axis
    
    def peak_detect(self):
        '''
        Performs peak detection on the fft magnitude result.
        I found a peak detection algorithm on github which I am using.
        I only detect peaks up to 4500kHz. I played with the "lookahead" 
        threshold somewhat to try and best pick out peaks. With more tweaking 
        this could be set better. Currently the lookahead sample number is 
        set to the number of samples for the peak detect (up to 4500kHz) 
        divided by 100.
        '''
        import peakdetect
        detect_samples = round(4500.0/self.delta_hertz)
        self.peaks = peakdetect.peakdetect(self.frames_fft_magn[0:detect_samples],freq_hertz[0:detect_samples],
                                      lookahead=round(detect_samples/50.0))
        self.peaks_max = array(self.peaks[:][:][0])
    
    def plot_annotated_spectrum(self):
        '''
        Calls the fft and peak detect routines. Then it plots the fft 
        magnitude spectrum in dB. I set the max freq to the highest note 
        I found (C8 at 4186.01Hz). This is needed because the highest 
        frequency calculated is the sampling rate of the music ~44kHz
        '''
        self.fft()
        self.peak_detect()
        #fft magn db max and min
        dBmax = self.frames_fft_magn_dB.max()
        dBmin = self.frames_fft_magn_dB.min()
        #call peak find routine
        fig = figure(figsize=(18,10))
        ax = fig.add_subplot(111)
        ax.set_title('Power Spectrum for {:s}. The red lines denote detected peaks with associated frequencies.'.format(self.sound_path))
        ax.plot(self.freq_hertz,self.frames_fft_magn_dB)
        ax.vlines(self.peaks_max[:,0],dBmin,dBmax,color='r')
        ax.set_xlabel('Frequency [Hz]')
        ax.set_ylabel('Power Amplitude [dB]')
        ax.set_xlim(0,4500)
        ax.set_ylim(dBmin,dBmax)
        for peak in self.peaks_max:
            ax.annotate('({:0.0f} Hz)'.format(peak[0]), 
                        xy=(peak[0],dBmax*0.9), xytext=(peak[0],dBmax*(0.75+0.2*rand())))
        show()
    
    def note2freq(self,note):
        '''
        Returns the frequency for a given note. Data converted to CSV from Wikipedia 
        http://en.wikipedia.org/wiki/Piano_key_frequencies
        '''
        notes_freq = loadtxt('notes_freq.txt',delimiter=',',dtype={'names': ('index', 'note', 'freq'),'formats': ('i4', 'S20', 'f8')})
        for line in notes_freq:
            if note==line[1]:
                return line[2]
    
    def print_noteandfreq(self):
        '''
        Prints the notes and freq from: http://en.wikipedia.org/wiki/Piano_key_frequencies
        '''
        notes_freq = loadtxt('notes_freq.txt',delimiter=',',dtype={'names': ('index', 'note', 'freq'),'formats': ('i4', 'S20', 'f8')})
        for line in notes_freq:
            print('{:2}) {:15s} {:7.2f} Hz'.format(line[0],line[1],line[2]))
```
