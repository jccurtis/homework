```python
from __future__ import print_function
import aifc
import struct
%pylab inline

test_aifc = aifc.open('sound_files/A4_PopOrgan.aif')
frame_rate = test_aifc.getframerate()
num_frames = test_aifc.getnframes() #
num_channels = test_aifc.getnchannels()
frame_width = test_aifc.getsampwidth() #number of frames per 32bit word
raw_frames = test_aifc.readframes(num_frames) #string of hex data
frame_dtype = {2:'uint16',4:'uint32'} #dictionary to map number of bytes per frames per 32 words to integer data type
frames_int = fromstring(raw_frames,dtype=frame_dtype[frame_width]) #numpy array of integer value of each frame
frames_float = frames_int.astype(float)
if num_channels == 2:
    frames_int_left = frames_int[::2] #left channel
    frames_int_right = frames_int[1::2] #right channel
    frames_int = frames_int_left+frames_int_right #sum of left and right channel for fft analysis

time = arange(size(frames_int)) / float(frame_rate) #array of timing for every time sample
```

```python
print('num_frames:   ',num_frames,'in the file')
print('frame_rate:   ',frame_rate,'Hz')
print('num_channels: ',num_channels,'(1 for mono, 2 for stereo)')
print('frame_width:  ',frame_width,'bytes')
print('length of sound: ',time[-1],'seconds')
```

```python
figure()
plot(time,frames_int)
xlabel("Time [s]")
ylabel("Amplitude")
title("Sound File")
```

```python
#Discrete Fourier Transform of Sound
frames_fft = fftpack.fft(frames_int)

#Polar magnitude of fft complex result
frames_fft_magn = sqrt(frames_fft.real**2+frames_fft.imag**2)

#Scaling the fft magnitude to dB relative to the minimum amplitude
frames_fft_magn_dB = 10*log10(frames_fft_magn/frames_fft_magn.min())

#Mapping frequency to samples:
#The spacing of the fft samples is equal to the inverse of the duration of the original discrete pulse. 
#With this freq spacing we create an array with the same length as the num of samples and proper 
#frequency spacing. This correlates the power of the fft to a frequency.
delta_hertz = 1/time[-1] #frequency spacing
freq_hertz = arange(0,len(frames_fft)*delta_hertz,delta_hertz) #frequency axis
```

```python
#I found a peak detection algorithm on github which I am using  
import peakdetect
peaks = peakdetect.peakdetect(frames_fft_magn_dB,freq_hertz,lookahead=3000)
peaks_max = array(peaks[:][:][0])
```

```python
#PLOTTING
#I set the max freq to the highest note I found (C8 at 4186.01Hz). 
#This is needed because the highest frequency calculated is the 
#sampling rate of the music ~44kHz
fig = figure(figsize=(15,15))
ax = fig.add_subplot(111)
ax.plot(freq_hertz,frames_fft_magn_dB)
ax.scatter(peaks_max[:,0],peaks_max[:,1], marker='x', c='r', s=500)
ax.set_ylim(frames_fft_magn_dB.min().real,frames_fft_magn_dB.max().real)
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Power Amplitude [dB]')
ax.set_xlim(0,4200)
show()
```

```python
#notes and frequencies from Wikipedia (http://en.wikipedia.org/wiki/Piano_key_frequencies):
notes_freq = loadtxt('notes_freq.txt',delimiter=',',dtype={'names': ('index', 'note', 'freq'),'formats': ('i4', 'S20', 'f8')})

def note2freq(note):
    for line in notes_freq:
        if note==line[1]:
            return line[2]
```
