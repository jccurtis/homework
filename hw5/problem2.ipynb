```python
from __future__ import print_function
import aifc
import struct
%pylab inline

#notes and frequencies from Wikipedia (http://en.wikipedia.org/wiki/Piano_key_frequencies):
notes_freq = loadtxt('notes_freq.txt',delimiter=',',dtype={'names': ('index', 'note', 'freq'),'formats': ('i4', 'S20', 'f8')})

test_aifc = aifc.open('sound_files/A4_PopOrgan.aif')
frame_rate = test_aifc.getframerate()
num_frames = test_aifc.getnframes() #
num_channels = test_aifc.getnchannels()
frame_width = test_aifc.getsampwidth() #number of frames per 32bit word
raw_frames = test_aifc.readframes(num_frames) #string of hex data
frame_dtype = {2:'uint16',4:'uint32'} #dictionary to map number of bytes per frames per 32 words to integer data type
frames_int = fromstring(raw_frames,dtype=frame_dtype[frame_width]) #numpy array of integer value of each frame
frames_float = frames_int.astype(float)
if num_channels == 2:
    frames_int_left = frames_int[::2]
    frames_int_right = frames_int[1::2]
time = arange(size(raw_frames)) / float(frame_rate)
```

```python
print('num_frames:   ',num_frames,'in the file')
print('frame_rate:   ',frame_rate,'Hz')
print('num_channels: ',num_channels,'(1 for mono, 2 for stereo)')
print('frame_width:  ',frame_width,'bytes')
```
